# Separating Target and Independent variables:

```python
x=df.drop("class",axis=1)
y=df["class"]
```

# Finding the Best Random State:

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report
```

```python
maxAccu=0
maxRS=0
for i in range(1,200):
	x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=.30, random_state=i)
	mod=GaussianNB()
	mod.fit(x_train,y_train)
	pred=mod.predict(x_test)
	acc=accuracy_score(y_test,pred)
	print(acc*100, "at random_state", i)
	if acc>maxAccu:
		maxAccu=acc
		maxRS=i
```

```python
print("Best accuracy is",maxAccu,"on Random_state",maxRS)
```

# Train Test Split:

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=.30, random_state=4)
```

# DecisionTreeClassifier:

```python
DTC= DecisionTreeClassifier()
DTC.fit(x_train,y_train)
pred=DTC.predict(x_test)
acc=classification_report(y_test,pred)
print(acc)
```

```python
from sklearn.metrics import roc_curve, auc
def roc_curve_acc(y_test, pred,method):
    false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, pred)
    roc_auc = auc(false_positive_rate, true_positive_rate)
    plt.title('Receiver Operating Characteristic')
    plt.plot(false_positive_rate, true_positive_rate, color='darkorange',label='%s AUC = %0.3f'%(method, roc_auc))
    plt.legend(loc='lower right')
    plt.plot([0,1],[0,1],'b--')
    plt.ylim([-0.1, 1.1])
    plt.xlim([-0.1, 1.1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
```

```python

roc_curve_acc(y_test, pred,"Decision Tree")
```
